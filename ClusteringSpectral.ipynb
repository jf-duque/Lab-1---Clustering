## 5.2 Selección numero de clusters

La selección del numero de clusters a usar se va a basar en en la forma grafica de como se vea el algorithmo, por esta razón en una primera instancia se va a realizar el un scatter plot para diferentes numero de clusters.

```python
dt_biciAlpes_transform= dt_biciAlpes[name_cols_number].copy()
selected= dt_biciAlpes[name_cols_number].copy()

#Transformación para el punto 6
# se 
dt_biciAlpes_transform = dt_biciAlpes_transform.dropna(subset = name_cols_number)
dt_biciAlpes_transform[['Accident_Severity'] + name_cols_number].isnull().sum() / dt_biciAlpes_transform.shape[0]
```

```python
def scatter_plot_cluster(data,
                    ylabel = 'Road surface condition',
                    xlabel = 'Accident severity',
                    title = ' Spectral clustering x num_clusters'):
  
    f = plt.figure(figsize=(15,10))
    f.add_subplot(3, 3, 1)

    for i in range(2, 9):
        sc = SpectralClustering(n_clusters=i,
                 n_init=10,
                 random_state=0).fit(data)
        f.add_subplot(3, 3, i-1)
        
        plt.scatter(data['Road_Surface_Conditions'], data['Accident_Severity'], c=sc.labels_, label="n_cluster-"+str(i), facecolor='red')
        plt.legend()
      

    plt.show() 
```

```python
scatter_plot_cluster(selected)
```

## 5.2 Conclusión Modelo 1

De las graficas anteriores se puede ver que los grupos al aumentar el numero de cllusters aumenta a cada una de las compinaciones que se pueden dar, realmente no es muy facil visualizar los grupos y no es muy facil tener una conclusión muy buena de la relación entre tipo de paimento y la gravedad de los accidentes.

```python
def bar_plot(data,
                    ylabel = 'Road surface condition',
                    xlabel = 'Accident severity',
                    title = ' Spectral clustering x num_clusters'):
    
    f = plt.figure(figsize=(15,10))
    f.add_subplot(3, 3, 1)

    for i in range(2, 9):
        sc = SpectralClustering(n_clusters=i,
                 n_init=10,
                 random_state=0).fit(data)
        f.add_subplot(3, 3, i-1)
        slected_bar = data.copy()
        labels = sc.labels_
        slected_bar['Cluster'] = labels
  
        
        cluster_distrib = slected_bar['Cluster'].value_counts()
        sns.barplot(x=cluster_distrib.index, y=cluster_distrib.values, color='b');

        
      

    plt.show()
```

```python
bar_plot(selected)
```
